Advance Visualization:(are done in order to compute the relationships between different columns)-Exploratory Data Analysis(EDA)
Work on:
1. Scatter plot:will help you to visualize two columns of your data(generate relationship between two columns data)-/part of matplotlip
2. Relational plot:help to visualize more than two columns data-help you to insight so many plot at a time/Part of saeborn
3. Box plot: part of catplot/ very statistical in nature
4. Pair plot:will pair the columns and it wil give u in different size
5. Headmap(basic of ML):
6. Catplot:refers to catogarical plot/ it has family/ as a family a lot of visualization insight. kind= strip,box,point,violin

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
data=pd.read_csv('/tips.csv')
```

```python
data.head(8)
```

```python
data.tail()
```

```python
data.plot.scatter(x='total_bill',y='tip',color='grey')# it is increasing between two column/ and the relation is linear in nature-with increase on total_bill, tip also increase.
```

# New Section

```python
snr.relplot(x='total_bill',y='tip',hue='sex',data=data)
#hue(name of parameter):hue is a color parameter/ any columns we put inside hue will become colourful
```

```python
snr.relplot(x='total_bill',y='tip',hue='sex',style='smoker',size='size',data=data)
# size will show according to the number of the size
# style is a parameter which give shape to the categories. Any column we put inside style, the categories inside will get a shape/
```

```python
snr.relplot(x='total_bill',y='tip',hue='sex',col='day',data=data)
# column: is a parameter which will give as numbers of plots same as the categories in your data.
```

```python
snr.relplot(x='total_bill',y='tip',hue='time',data=data)
```

```python
snr.relplot(x='total_bill',y='tip',col='time',data=data,color='g')
```

```python
snr.pairplot(data)
```

```python
snr.catplot(x='day',y='total_bill',hue='sex',data=data)
```

```python
snr.catplot(x='day',y='total_bill',hue='sex',kind='bar',data=data)
#kind= strip,point,voilent
```

```python
snr.catplot(x='total_bill',y='tip',hue='sex',kind='violin',data=data)
```

```python
snr.catplot(x='day',y='total_bill',hue='sex',kind='box',data=data)
```

```python
snr.catplot(x='day',y='total_bill',hue='sex',kind='point',data=data)
```
